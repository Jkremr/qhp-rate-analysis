```python
import numpy as np
import pandas as pd
```

```python
pd.options.display.max_seq_items = 2000
```

```python
plan_rates_2014 = pd.read_csv('./data_files/Plan_Rates/Rate_PUF_2014.csv',encoding = "ISO-8859-1" )
plan_rates_2015 = pd.read_csv('./data_files/Plan_Rates/Rate_PUF_2015.csv',encoding = "ISO-8859-1" )
plan_rates_2016 = pd.read_csv('./data_files/Plan_Rates/Rate_PUF_2016.csv',encoding = "ISO-8859-1" )
plan_rates_2017 = pd.read_csv('./data_files/Plan_Rates/Rate_PUF_2017.csv',encoding = "ISO-8859-1" )
plan_rates_2018 = pd.read_csv('./data_files/Plan_Rates/Rate_PUF_2018.csv',encoding = "ISO-8859-1" )

plan_rates_2014.name = 'plan_rates_2014_PUF'
plan_rates_2015.name = 'plan_rates_2015_PUF'
plan_rates_2016.name = 'plan_rates_2016_PUF'
plan_rates_2017.name = 'plan_rates_2017_PUF'
plan_rates_2018.name = 'plan_rates_2018_PUF'
```

```python
dfs = [plan_rates_2014, plan_rates_2015, plan_rates_2016, plan_rates_2017, plan_rates_2018]
```

```python
for frame in dfs:
    print("%r has %r columns and %r rows." % (frame.name, len(list(frame)), len(frame)))
```

```python
n = 0
for frame in dfs:
    n += len(frame)
print("For a total of %r rows." % (n))
```

```python
consolidated_plan_rates = pd.DataFrame()
for frame in dfs:
    consolidated_plan_rates = consolidated_plan_rates.append(frame)

consolidated_plan_rates.name = 'consolidated_plan_rates_2014-2018'

    
print("%r has %r columns and %r rows." % (consolidated_plan_rates.name, len(list(consolidated_plan_rates)), len(consolidated_plan_rates)))
```

```python
len(list(consolidated_plan_rates))
```

```python
consolidated_plan_rates.dtypes
```

```python
float_to_int_col_names = [
    'IssuerId2',
    'VersionNum'
]

consolidated_plan_rates[float_to_int_col_names] = consolidated_plan_rates[float_to_int_col_names].apply(np.int64)
consolidated_plan_rates[float_to_int_col_names]
```

```python
for column in consolidated_plan_rates:
    print("Column %r has values of: %r" % (consolidated_plan_rates[column].name ,consolidated_plan_rates[column].unique()))
```

```python
consolidated_plan_rates.columns = [c.lower() for c in consolidated_plan_rates.columns] #postgres doesn't like capitals or spaces
```

```python
rates_2014 = consolidated_plan_rates.loc[consolidated_plan_rates['businessyear'] == 2014]
rates_2015 = consolidated_plan_rates.loc[consolidated_plan_rates['businessyear'] == 2015]
rates_2016 = consolidated_plan_rates.loc[consolidated_plan_rates['businessyear'] == 2016]
rates_2017 = consolidated_plan_rates.loc[consolidated_plan_rates['businessyear'] == 2017]
rates_2018 = consolidated_plan_rates.loc[consolidated_plan_rates['businessyear'] == 2018]

```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://localhost:5432/ifprates')
```

```python
rates_2014.to_sql("plan_rates_2014_etl", engine,)
```

```python
rates_2015.to_sql("plan_rates_2015_etl", engine)
```
