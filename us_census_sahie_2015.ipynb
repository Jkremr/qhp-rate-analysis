```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data_files/us_census_sahie_2015.csv',encoding = "ISO-8859-1" )
df.head()
```

```python
df[['statefips', 'countyfips']] = df[['statefips', 'countyfips']].astype(str) #apply(pd.to_string, errors='raise')

def county_fips_complete(x):
    if len(x) == 1:
        x = '00' + x
    elif len(x) == 2:
        x = '0' + x
    else:
        x = x
    return x

def state_fips_complete(x):
    if len(x) == 1:
        x = '0' + x
    else:
        x = x
    return x
    
df['countyfips'] = df['countyfips'].apply(lambda x: county_fips_complete(x))
df['statefips'] = df['statefips'].apply(lambda x: state_fips_complete(x))

df['fips_code'] = df[['statefips', 'countyfips']].apply(lambda x: ''.join(x), axis=1)


df.head()
```

```python
df = df[['year', 'fips_code', 'geocat','agecat','racecat', 'sexcat','iprcat','NIPR','NUI','NIC', 'county_name', 'state_name']]

col_rename = {
    'geocat': 'geo_category_id',
    'agecat': 'age_category_id',
    'racecat': 'race_category_id',
    'sexcat': 'sex_category_id',
    'iprcat': 'income_category_id',
    'NIPR': 'total_in_group',
    'NUI': 'number_uninsured',
    'NIC': 'number_insured'

}

df.rename(columns=col_rename, inplace = True)
```

```python
geo_category_names = {
    40: 'State Identifier',
    50: 'County Identifier'
}

age_category_names = {
    0: 'Under 65 years',
    1: '18 to 64 years',
    2: '40 to 64 years',
    3: '50 to 64 years',
    4: 'Under 19 years',
    5: '21 to 64 years'
}

race_category_names = {
    0: 'All races',
    1: 'White alone not Hispanic',
    2: 'Black alone not Hispanic',
    3: 'Hispanic (any race)'
}

sex_category_names = {
    0: 'Both',
    1: 'Male',
    2: 'Female'
}

income_category_names = {
    0: 'All income levels',
    1: 'At or below 200% of poverty',
    2: 'At or below 250% of poverty',
    3: 'At or below 138% of poverty',
    4: 'At or below 400% of poverty',
    5: 'Between 138% - 400%  of poverty'
}

df['geo_category_names'] = df['geo_category_id'].map(lambda x: geo_category_names[x])
df['age_category_names'] = df['age_category_id'].map(lambda x: age_category_names[x])
df['race_category_names'] = df['race_category_id'].map(lambda x: race_category_names[x])
df['sex_category_names'] = df['sex_category_id'].map(lambda x: sex_category_names[x])
df['income_category_names'] = df['income_category_id'].map(lambda x: income_category_names[x])
```

```python
df[['total_in_group','number_uninsured','number_insured', 'fips_code', 'year']] =\
    df[['total_in_group','number_uninsured','number_insured', 'fips_code', 'year']].apply(pd.to_numeric, errors='coerce', downcast='integer')

df.head()
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://localhost:5432/ifprates')
```

```python
df.columns = [c.lower() for c in df.columns] #postgres doesn't like capitals or spaces
```

```python
df.to_sql("census_sahie_2015", engine, schema='etl')
```
